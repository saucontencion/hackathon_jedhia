```python
import gdown
import pandas as pd

# Reemplaza 'ID_DEL_ARCHIVO' con el ID real del archivo
file_id = "1DHbyGM83tjwRJ3yTFF-ARz5VoZh9Cfu5"
url = f"https://drive.google.com/uc?id={file_id}"
output = "dataset.xlsx"

# Descargar el archivo
gdown.download(url, output, quiet=False)

# Cargar el archivo en un DataFrame
df = pd.read_excel(output)

# Verifica la carga de los datos
print(df.head())
```

```python
# Seleccionar la columna 'paises' y convertirlo a lista para poder tokenizarlo
list_occupations = df['ocupación'].astype(str).tolist()  # Convertimos a una lista de strings por si hay valores no string
print(list_occupations)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Cargar el modelo de JINA y el tokenizer desde Hugging Face
model_name = "jinaai/jina-embeddings-v3"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModel.from_pretrained(model_name, trust_remote_code=True)
```

```python
# Tokenizar la lista de paises
dictionary_countries = {}
for value in list_countries:
    inputs = tokenizer(value, padding=True, truncation=True, return_tensors="pt")
    # Generar los embeddings
    with torch.no_grad():
        outputs = model(**inputs)
        dictionary_countries[value] = outputs.last_hidden_state.mean(dim=1)  # Promediar a lo largo de la secuencia para obtener un vector único por texto

# Ver los embeddings resultantes
print(dictionary_countries)
```

```python
# Tokenizar la lista de paises
inputs = tokenizer(list_occupations, padding=True, truncation=True, return_tensors="pt")

# Generar los embeddings
with torch.no_grad():
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1)  # Promediar a lo largo de la secuencia para obtener un vector único por texto

# Ver los embeddings resultantes
print(embeddings)
```

```python
import umap
import matplotlib.pyplot as plt

# Supongamos que embeddings es tu matriz de embedding, de dimensiones (n_samples, n_features)

# Reducir la dimensionalidad con UMAP
umap_model = umap.UMAP(n_components=2, random_state=42)  # Para 2D
embeddings_2d = umap_model.fit_transform(embeddings)

# Visualización en 2D
plt.figure(figsize=(20, 15))
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], s=10, alpha=0.7)

# Añadir glosas a cada punto
for i, label in enumerate(list_countries):
    plt.text(embeddings_2d[i, 0], embeddings_2d[i, 1], label[:20], fontsize=8, alpha=0.7)

plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.title("UMAP de Embeddings")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Usar DBSCAN para agrupamiento basado en densidad
db = DBSCAN(eps=0.2, min_samples=1, metric='cosine')  # Ajusta los parámetros según sea necesario
db.fit(embeddings)

# Mostrar los países agrupados
df_dbscan = pd.DataFrame({
    'pais': list_occupations,
    'cluster': db.labels_
})

print(df_dbscan.sort_values(by='cluster'))
```
