```python
import gdown
import pandas as pd

# Reemplaza 'ID_DEL_ARCHIVO' con el ID real del archivo
file_id = "1DHbyGM83tjwRJ3yTFF-ARz5VoZh9Cfu5"
url = f"https://drive.google.com/uc?id={file_id}"
output = "dataset.xlsx"

# Descargar el archivo
gdown.download(url, output, quiet=False)

# Cargar el archivo en un DataFrame
df = pd.read_excel(output)

# Verifica la carga de los datos
print(df.head())
```

```python
# Seleccionar la columna 'paises' y convertirlo a lista para poder tokenizarlo
list_countries = df['pais'].astype(str).tolist()  # Convertimos a una lista de strings por si hay valores no string
print(list_countries)
```

```python
!pip install transformers torch
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Cargar el modelo de JINA y el tokenizer desde Hugging Face
model_name = "jinaai/jina-embeddings-v3"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
# Tokenizar la lista de paises
inputs = tokenizer(list_countries, padding=True, truncation=True, return_tensors="pt")

# Generar los embeddings
with torch.no_grad():
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1)  # Promediar a lo largo de la secuencia para obtener un vector único por texto

# Ver los embeddings resultantes
print(embeddings)
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Usar DBSCAN para agrupamiento basado en densidad
db = DBSCAN(eps=0.2, min_samples=1, metric='cosine')  # Ajusta los parámetros según sea necesario
db.fit(embeddings)

# Mostrar los países agrupados
df_dbscan = pd.DataFrame({
    'pais': list_countries,
    'cluster': db.labels_
})

print(df_dbscan.sort_values(by='cluster'))
```

```python
import matplotlib.pyplot as plt

# Agrupamos por el cluster y contamos la cantidad de elementos en cada uno
cluster_counts = df_dbscan.groupby('cluster').size()

# Tomamos el nombre del primer elemento de cada cluster para usar como etiquetas
cluster_names = df_dbscan.groupby('cluster')['pais'].first().str[:3]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(cluster_names, cluster_counts)

# Etiquetas y título
plt.xlabel("Paises")
plt.ylabel("Cantidad según pais")
plt.title("Grafico por paises")

# Añadir la cantidad de elementos encima de cada barra
for bar in bars:
    yval = bar.get_height()  # Obtener la altura de la barra (cantidad de elementos)
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval),
             ha='center', va='bottom')  # Colocar el texto en el centro y encima de la barra


plt.show()
```
